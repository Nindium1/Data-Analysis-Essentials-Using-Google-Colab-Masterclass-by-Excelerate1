#Lists

```python
monthly_expenses = [2500, 3000, 2800, 3200, 2900]
```

```python
print (f"total_expenses:{sum(monthly_expenses)}")
print (f"average_expense:{sum(monthly_expenses)/len(monthly_expenses)}")
print (f"max_expense:{max(monthly_expenses)}")
print (f"min_expense: {min(monthly_expenses)}")
```

#Conditionals

```python
def calculate_discount(purchase_amount):
  if purchase_amount > 500:
    print("eligible for 20% discount")
  elif purchase_amount > 100:
    print("eligible for 10% discount")
  else:
    print("no discount")
```

```python
print("Test 1:")
calculate_discount(550)
```

```python
print("\nTest 2:")
calculate_discount(175)
```

```python
print("\nTest 3:")
calculate_discount(75)
```

#Pandas Dataframe Creation and Manipulation

```python
import pandas as pd
student_data={
    'Name': ['Alice', 'Charles', 'Bob', 'David'],
    'Math_Score': [85,92,78,88],
    'Science_Score': [90,88,85,95],
    'English_Score':[75,80,82,79]

}
```

```python
df=pd.DataFrame(student_data)
df
```

```python
df['Average_Score'] = (df['Math_Score'] +
                                 df['Science_Score'] +
                                 df['English_Score']) / 3
```

```python
df
```

```python
Smartest_Students = df[df['Average_Score'] > 85]
```

```python
print ("Smartest Students:")
Smartest_Students
```

#Numpy Arrray Operations

```python
import numpy as np
product_a_sales = np.array([120, 150, 130, 180, 160])
product_b_sales = np.array([100, 130, 140, 170, 150])
```

```python
Total_sales=product_a_sales + product_b_sales
print ("Total_sales:")
Total_sales
```

```python
Difference_in_sales=product_a_sales - product_b_sales
print ("Difference_in_sales:")
Difference_in_sales
```

```python
average_sales_a = np.mean(product_a_sales)
average_sales_b = np.mean(product_b_sales)
print(f"Average Sales for Product A: {average_sales_a}")
print(f"Average Sales for Product B: {average_sales_b}")
```

```python
is_a_greater = product_a_sales > product_b_sales
print("Is Product A Sales Greater than Product B?")
print(is_a_greater)
```

```python
count_a_greater = np.sum(is_a_greater)
print(f"Number of months Product A sales were greater: {count_a_greater}")
```

#MatPlotLib Basic Plotting

```python
import matplotlib.pyplot as plt
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 5, 4, 5]
```

```python
plt.plot(x_values, y_values, label="Data Series")
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
```

#Seaborn Basic Plotting

```python
import seaborn as sns
category_data = pd.DataFrame({
    'Category': ['Type A', 'Type B', 'Type A', 'Type C', 'Type B', 'Type C'],
    'Value': [10, 15, 12, 8, 18, 10]
})
```

```python
sns.countplot(data=category_data, x='Category')
plt.title("Count of Each Category")
```

```python
sns.barplot(data=category_data, x='Category', y='Value')
plt.title("Average Value by Category")
plt.show()
```

```python
correlation_matrix = np.array([
    [1.0, 0.8, 0.3],
    [0.8, 1.0, 0.5],
    [0.3, 0.5, 1.0]
])
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix Heatmap")
```
